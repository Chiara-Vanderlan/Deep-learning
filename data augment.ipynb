```python
x= img_to_array(img)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator (
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
x = x.reshape((1,) + x.shape)
```

```python
x.shape
```

```python
i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir='aug_images', save_prefix='elephant', save_format='jpg' ):
  i+=1
  if i>10:
    break
```
